```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = [
    "I love machine learning",
    "Machine learning is amazing",
    "I love deep learning",
    "Deep learning drives AI"
]
```

```python
#create a bag of words represenatation of text data..
vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform(documents)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("BoW Matrix:\n", bow_matrix.toarray())
```

```python
# Generate TF-IDF matrix using sklearn
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python
#compare BoW-TF_IDF outputs
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

print("BoW:\n", bow_df)
print("\nTF-IDF:\n", tfidf_df)
```

```python
#calculate cosine similarity between two documents 
cos_sim = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])
print("Cosine Similarity:", cos_sim[0][0])
```

```python
#Demonstrating semantic and similarity using word vector 
doc1 = ["machine learning is powerful"]
doc2 = ["AI and machine learning are closely related"]

tfidf = TfidfVectorizer()
vectors = tfidf.fit_transform(doc1 + doc2)

similarity = cosine_similarity(vectors[0], vectors[1])

print("Semantic Similarity:", similarity[0][0])
```

```python
#Build a mini text processing pipeline combining all steps.
def text_pipeline(docs):
    bow = CountVectorizer()
    bow_matrix = bow.fit_transform(docs)
    tfidf = TfidfVectorizer()
    tfidf_matrix = tfidf.fit_transform(docs)
    similarity = cosine_similarity(tfidf_matrix)
    return {
        "BoW": bow_matrix.toarray(),
        "TF-IDF": tfidf_matrix.toarray(),
        "Similarity": similarity
    }

results = text_pipeline(documents)

print("BoW:\n", results["BoW"])
print("\nTF-IDF:\n", results["TF-IDF"])
print("\nSimilarity:\n", results["Similarity"])
```
