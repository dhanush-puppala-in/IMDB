```python
##unigram generation
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
sentence = "This is a sample sentence for generating bigrams."
tokens = word_tokenize(sentence)
unigrams = list(ngrams(tokens, 1))
print(unigrams)
```

```python
##bigram generation
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
sentence = "This is a sample sentence for generating bigrams."
tokens = word_tokenize(sentence)
bigrams = list(ngrams(tokens, 2))
print(bigrams)
```

```python
from collections import Counter
from nltk.corpus import stopwords
text = "This is a sample sentence, showing off the stop words filtration."
token = word_tokenize('english')
st = set(stopwords.words('english'))
tokens = [word for word in token if word not in st]
tk = []
for w in tokens:
    tk.append(w)
freq = Counter(tokens)
print(Counter(tk))
```

```python
import matplotlib.pyplot as plt
label = list(freq.keys())
plt.bar(label, freq.values())
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Frequency Distribution')
plt.xticks(rotation=45)
plt.show()  
```

```python
import nltk
import re
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
nltk.download('punkt')
text = "the product is amazying, but,  the delivery was dely"
text = text.lower()                     
text = re.sub(r'[^a-z\s]', '', text)    
text = re.sub(r'\s+', ' ', text).strip() 
print("Cleaned Text:", text)
tokens = word_tokenize(text)
print("Tokens:", tokens)
unigrams = list(ngrams(tokens, 1))
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))
print("\nUnigrams:")
print(unigrams)
print("\nBigrams:")
print(bigrams)
print("\nTrigrams:")
print(trigrams)
```
